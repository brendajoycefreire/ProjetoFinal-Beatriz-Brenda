## Iniciando o tratamento e limpeza de dados da base intitulada "impactos do trabalho remoto na saúde mental" disponibilizado pelo site kaggle

```python
import pandas as pd
```

```python
df = pd.read_csv(r'Impact_of_Remote_Work_on_Mental_Health.csv')
df.head()
```

### Após importar a biblioteca pandas e fazer a leitura da base de dados, foi utilizado o head para se certificar de que o dataset foi carregado de forma correta

```python
df.describe(include='all')
```

```python
df.info()
```

### Pedimos uma leitura estatística geral do dataset analisado (describe) e solicitamos ao dataframe as informações da base para observar a ocorrência de dados nulos e validar os tipos (dtype) de cada coluna. Após observação das informações retornadas, decidimos por filtrar as colunas mais relevantes para a nossa análise, excluindo 6 das 19 colunas que anteriormente existiam.

```python
df = df.drop(['Job_Role', 'Years_of_Experience', 'Company_Support_for_Remote_Work', 'Physical_Activity', 'Region'], axis=1)
df.info()
```

```python
df.describe(include='all')
```

### Após finalizado o tratamento dos dados, um novo arquivo csv foi gerado para ser utilizado nas análises futuras

```python
df.to_csv('Dados_SaudeMental.csv', index=False)
```

## Iniciando as análises do dataset tratado

```python
import matplotlib.pyplot as plt
```

### Para gerar os gráficos deste trabalho será utilizada a biblioteca do matplotlib

```python
modalidade = df['Work_Location'].value_counts()

# criação do gráfico

modalidade.plot(kind="bar", edgecolor="blue", color="pink")
barras = modalidade.plot.bar(color='pink')

# configurações
plt.xlabel("Modalidade de Trabalho")
plt.xticks(rotation=0)
plt.ylabel("Quantidade")
plt.title("Modalidade de Trabalho dos Colaboradores")
plt.legend

# Adicionar rótulos nos graficos
for i, v in enumerate(modalidade):
    barras.text(i, v + 0.01, f'{v}', color='black', ha='center')

plt.show()


print('Notamos que mais da metade da base trabalha de forma remota ou hibrida.')
```

```python
estresse = df['Stress_Level'].value_counts()

# criação do gráfico

estresse.plot(kind="bar", edgecolor="blue", color="pink")
barras = estresse.plot.bar(color='pink')

# configurações
plt.xlabel("Nivel de estresse dos colaboradores")
plt.xticks(rotation=0)
plt.ylabel("Quantidade")
plt.title("Nivel de estresse dos Colaboradores")
plt.legend

# Adicionar rótulos nos graficos
for i, v in enumerate(estresse):
    barras.text(i, v + 0.01, f'{v}', color='black', ha='center')

plt.show()


print('Notamos que a maioria dos colaboradores estão com nivel de estresse alto.')
```

```python
saudeMental = df['Mental_Health_Condition'].value_counts()

# criação do gráfico

saudeMental.plot(kind="bar", edgecolor="blue", color="pink")
barras = saudeMental.plot.bar(color='pink')

# configurações
plt.xlabel("Saude Mental dos colaboradores")
plt.xticks(rotation=0)
plt.ylabel("Quantidade")
plt.title("Saude Mental dos Colaboradores")
plt.legend

# Adicionar rótulos nos graficos
for i, v in enumerate(saudeMental):
    barras.text(i, v + 0.01, f'{v}', color='black', ha='center')

plt.show()


print('Notamos que a maioria dos colaboradores está dividida em burnout e ansiedade.')
```
